```python
import pandas as pd
```

```python
osha_df = pd.read_excel("osha_inspection_2016-2023.xlsx")
osha_df.head()
```

```python
recert_df = pd.read_excel("filtered_recert(recertified).xlsx")
recert_df.head()
```

```python
vpp_df = pd.read_excel("filtered_file(vpp).xlsx")
vpp_df.head()
```

```python
vpp_df.rename(
    columns={'city': 'city_vpp',
             'state': 'state_vpp'},
               inplace=True)
```

```python
recert_df.rename(
    columns={'city': 'city_recert',
             'state': 'state_recert'},
    inplace=True)
```

Data Cleaning

```python
osha_df['estab_name'] = osha_df['estab_name'].str.strip(
).str.lower()
osha_df['site_city'] = osha_df['site_city'].str.strip().str.lower()
osha_df['site_state'] = osha_df['site_state'].str.strip().str.lower()
```

```python
vpp_df['establishment_name'] = vpp_df['establishment_name'] .str.strip(
).str.lower()
vpp_df['city_vpp'] = vpp_df['city_vpp'].str.strip().str.lower()
vpp_df['state_vpp'] = vpp_df['state_vpp'].str.strip().str.lower()
```

```python
recert_df['establishment_name'] = recert_df['establishment_name'] .str.strip(
).str.lower()
recert_df['city_recert'] = recert_df['city_recert'].str.strip().str.lower()
recert_df['state_recert'] = recert_df['state_recert'].str.strip().str.lower()
```

```python
vpp_osha_results = pd.merge(osha_df, vpp_df,
                            left_on=['estab_name', 'site_city',
                                     'site_state', 'naics_code'],
                            right_on=['establishment_name', 'city_vpp', 'state_vpp', 'NAICs'])
vpp_osha_results.head()
```

```python
vpp_osha_results.to_excel("new_vpp_osha_df.xlsx", index=False)
```

```python
recert_osha_results = pd.merge(osha_df, recert_df,
                               left_on=['estab_name', 'site_city',
                                        'site_state', 'naics_code'],
                               right_on=['establishment_name', 'city_recert', 'state_recert', 'NAICs'])
recert_osha_results.head()
```

```python
recert_osha_results.to_excel("new_recert_df.xlsx", index=False)
```
