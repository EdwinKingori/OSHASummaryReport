```python
import pandas as pd
```

```python
osha_df = pd.read_excel("osha_inspection_2016-2023.xlsx")
osha_df.head()
```

```python
df_2018 = pd.read_csv(
    "/home/eddy/Documents/OSHA_Injury_data_analysis/merged_dfs/ITA Data CY 2018.csv", encoding='ISO-8859-1')
df_2018.head()
```

```python
osha_df['estab_name'] = osha_df['estab_name'].str.strip(
).str.lower()
osha_df['site_city'] = osha_df['site_city'].str.strip().str.lower()
osha_df['site_state'] = osha_df['site_state'].str.strip().str.lower()
```

```python
df_2018['establishment_name'] = df_2018['establishment_name'].str.strip(
).str.lower()
df_2018['city'] = df_2018['city'].str.strip().str.lower()
df_2018['state'] = df_2018['state'].str.strip().str.lower()
```

```python
df_2018 = df_2018.rename(columns={'naics_code': 'naics_code_2018'})
```

```python
merging_results = pd.merge(osha_df, df_2018,
                           left_on=['estab_name', 'site_city',
                                    'site_state', 'naics_code'],
                           right_on=['establishment_name', 'city', 'state', 'naics_code_2018'],
                           suffixes=['_osha','_df'])
merging_results.head()
```

```python
merging_results.columns 
```

```python
merging_results.to_csv("missing_2018_data.csv", index=False)
```

Merging osha_data with vpp and recertified datasets.

```python
osha_df.head()
```

```python
new_vpp_df = pd.read_excel("NEW VPP.xlsx")
new_vpp_df.head()
```

```python
recert_vpp_df = pd.read_excel("RECERT VPP.xlsx")
recert_vpp_df.head()
```

Cleaning data columns

```python
new_vpp_df['company_name'] = new_vpp_df['company_name'] .str.strip(
).str.lower()
new_vpp_df['city'] = new_vpp_df['city'].str.strip().str.lower()
new_vpp_df['state'] = new_vpp_df['state'].str.strip().str.lower()
```

```python
recert_vpp_df['company_name'] = recert_vpp_df['company_name'] .str.strip(
).str.lower()
recert_vpp_df['city'] = recert_vpp_df['city'].str.strip().str.lower()
recert_vpp_df['state'] = recert_vpp_df['state'].str.strip().str.lower()
```

Merging data

```python
vpp_osha_results = pd.merge(osha_df, new_vpp_df,
                           left_on=['estab_name', 'site_city',
                                    'site_state'],
                   right_on=['company_name', 'city',
                                     'state'])
vpp_osha_results.head()
```

```python
vpp_osha_results.to_csv("vpp_osha_data.csv", index=False)
```

```python
recertified_osha_results = pd.merge(osha_df, recert_vpp_df,
                            left_on=['estab_name', 'site_city',
                                     'site_state'],
                            right_on=['company_name', 'city',
                                      'state'])
recertified_osha_results.head()
```

```python
recertified_osha_results.to_csv("recertified_osha_data.csv", index=False)
```
