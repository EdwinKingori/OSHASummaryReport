```python
import pandas as pd
```

Importing datasets

```python
osha_df = pd.read_excel("osha_inspection_2016-2023.xlsx")
osha_df.head()
```

```python
vpp_df = pd.read_excel("vpp_merged(revised).xlsx")
vpp_df.head()
```

```python
recert_df = pd.read_excel("recertified_merged(revised).xlsx")
recert_df.head()
```

Re-formattting the columns

```python
osha_df['estab_name'] = osha_df['estab_name'].str.strip(
).str.lower()
osha_df['site_city'] = osha_df['site_city'].str.strip().str.lower()
osha_df['site_state'] = osha_df['site_state'].str.strip().str.lower()
```

```python
vpp_df['establishment_name_vpp'] = vpp_df['establishment_name_vpp'] .str.strip(
).str.lower()
vpp_df['city_vpp'] = vpp_df['city_vpp'].str.strip().str.lower()
vpp_df['state_vpp'] = vpp_df['state_vpp'].str.strip().str.lower() 
```

```python
recert_df['establishment_name_recert'] = recert_df['establishment_name_recert'] .str.strip(
).str.lower()
recert_df['city_vpp'] = recert_df['city_vpp'].str.strip().str.lower()
recert_df['state_vpp'] = recert_df['state_vpp'].str.strip().str.lower()
```

```python
vpp_df.rename(
    columns={'naics_code': 'naics_code_vpp'}, inplace=True)

recert_df.rename(
    columns={'naics_code': 'naics_code_recert'}, inplace=True)
```

```python
vpp_osha_results = pd.merge(osha_df, vpp_df,
                            left_on=['estab_name', 'site_city', 'site_state', 'naics_code'],
                            right_on=['establishment_name_vpp', 'city_vpp', 'state_vpp', 'naics_code_vpp'])
vpp_osha_results.head(5)
```

```python
vpp_osha_results.to_csv("test_csv(1).csv", index=False)
```

```python
recert_osha_results = pd.merge(osha_df, recert_df,
                            left_on=['estab_name', 'site_city',
                                     'site_state', 'naics_code'],
                            right_on=['establishment_name_recert', 'city_vpp', 'state_vpp', 'naics_code_recert'])
recert_osha_results.head(5)
```

```python
recert_osha_results.to_csv("recert_merged_osha.csv", index=False)
```

```python
vpp_osha_null_results = pd.merge(osha_df, vpp_df,
                            left_on=['estab_name', 'site_city', 'site_state', 'naics_code'],
                            right_on=['establishment_name_vpp', 'city_vpp', 'state_vpp', 'naics_code_vpp'],
                            how='left')
vpp_osha_null_results.head(5)
```

```python
vpp_osha_null_results.to_csv("vpp_merge_osha_nulls_csv", index=False)
```

```python
recert_osha_results_null = pd.merge(osha_df, recert_df,
                            left_on=['estab_name', 'site_city',
                                     'site_state', 'naics_code'],
                                    right_on=[
                                        'establishment_name_recert', 'city_vpp', 'state_vpp', 'naics_code_recert'],
                                    how="left")
recert_osha_results_null.head(5)
```

```python
recert_osha_results_null.to_csv("recert_merged_osha_null.csv", index=False)
```
