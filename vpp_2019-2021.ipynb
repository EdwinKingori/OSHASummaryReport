```python
import pandas as pd
```

Importing data tables

```python
summary_df = pd.read_csv("Revised_merged_summary.csv")
summary_df.columns
```

2019 Non_Construction_Site_analysis

```python
df_2019 = pd.read_csv("3-VPP Annual Rates 2019 - Non-Construction Sites.csv")
df_2019.head()
```

```python
df_2019['site_name'] = df_2019['site_name'] .str.strip(
).str.lower()
df_2019['City'] = df_2019['City'].str.strip().str.lower()
df_2019['State'] = df_2019['State'].str.strip().str.lower()
```

```python
summary_df['establishment_name'] = summary_df['establishment_name'] .str.strip(
).str.lower()
summary_df['city'] = summary_df['city'].str.strip().str.lower()
summary_df['state'] = summary_df['state'].str.strip().str.lower()
```

```python
vpp_2019_find = pd.merge(summary_df, df_2019,
                         left_on=['establishment_name', 'city','state'],
                         right_on=['site_name', 'City','State'])
vpp_2019_find.head(5)
```

```python
vpp_2019_find.to_csv("vpp_2019_df.csv", index=False)
```
